```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
listingdata = pd.read_csv('listings.csv')
listingdata.head()
```

```python
host_verify = pd.DataFrame(listingdata['host_verifications'])
host_profilePic = pd.DataFrame(listingdata['host_has_profile_pic'])
host_idVerify = pd.DataFrame(listingdata['host_identity_verified'])
reviewVal = pd.DataFrame(listingdata['review_scores_value'])
noOfReviews = pd.DataFrame(listingdata['number_of_reviews'])
```

```python
host_verify.info()
host_profilePic.info()
host_idVerify.info()
reviewVal.info()
noOfReviews.info()
```

```python
# Create a copy of the Dataset
listingdata_clean = listingdata.copy()

# Rename
listingdata_clean.rename(columns = {'host_verifications': 'verification'}, inplace = True)
listingdata_clean.rename(columns = {'host_has_profile_pic': 'hasProfilePic'}, inplace = True)
listingdata_clean.rename(columns = {'host_identity_verified': 'identityVerified'}, inplace = True)
listingdata_clean.rename(columns = {'review_scores_value': 'reviewScore'}, inplace = True)
listingdata_clean.rename(columns = {'number_of_reviews': 'noOfReviews'}, inplace = True)

# Fill missing values in data
listingdata_clean["verification"].fillna(value = "NoType", inplace = True)
listingdata_clean["hasProfilePic"].fillna(value = "NoType", inplace = True)
listingdata_clean["identityVerified"].fillna(value = "NoType", inplace = True)
```
